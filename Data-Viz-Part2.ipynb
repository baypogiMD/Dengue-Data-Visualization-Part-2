```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```
Philippines = pd.read_csv('dengue_climate.csv')
```

```
f, ax = plt.subplots(figsize=(20, 4), constrained_layout=True)
sns.violinplot(y="Temperature", x="Region", data=Philippines, palette="Spectral", inner="quart", bw =0.5, cut=1, linewidth=1)
ax.set_xlabel("\n Administrative Regions", size=16)
ax.set_ylabel("Temperature (C) \n", size=16)
f.suptitle("Monthly temperature in each administrative regions in the Philippines from 2013 to 2018", size=20)

plt.savefig('temp.jpg', dpi=150)
plt.show()
```

```
f, ax = plt.subplots(figsize=(20, 4), constrained_layout=True)
sns.violinplot(y="Humidity", x="Region", data=Philippines, palette="Spectral", inner="quart", bw =0.2, cut=1, linewidth=1)
ax.set_xlabel("\n Administrative Regions", size=16)
ax.set_ylabel("Humidity (%) \n", size=16)
f.suptitle("Monthly humidity in each administrative regions in the Philippines from 2013 to 2018", size=20)

plt.savefig('humid.jpg', dpi=150)
plt.show()
```

```
f, ax = plt.subplots(figsize=(20, 4), constrained_layout=True)
sns.violinplot(y="Rainfall", x="Region", data=Philippines, palette="Spectral", inner="quart", bw =0.2, cut=1, linewidth=1)
ax.set_xlabel("\n Administrative Regions", size=16)
ax.set_ylabel("Rainfall (mm) \n", size=16)
f.suptitle("Monthly rainfall in each administrative regions in the Philippines from 2013 to 2018", size=20)

plt.savefig('rain.jpg', dpi=150)
plt.show()
```

```
f, ax = plt.subplots(figsize=(20, 4), constrained_layout=True)
sns.violinplot(y="Dengue_Count", x="Region", data=Philippines, palette="colorblind", inner="quart", bw = 0.2, cut=1, linewidth=1)
ax.set_xlabel("\n Administrative Regions", size=16)
ax.set_ylabel("Dengue Incidence \n", size=16)
f.suptitle("Monthly dengue incidence in each administrative regions in the Philippines from 2013 to 2018", size=20)

plt.savefig('dengue1.jpg', dpi=150)
plt.show()
```

```
Region = Philippines['Region']
Dengue_Count = np.log(Philippines.Dengue_Count +1)
Dengue_Count = pd.DataFrame(Dengue_Count, columns=['Dengue_Count'])
Philippines2 = pd.concat([Region, Dengue_Count], axis=1)
```

```
f, ax = plt.subplots(figsize=(20, 4), constrained_layout=True)
sns.violinplot(y="Dengue_Count", x="Region", data=Philippines2, palette="Spectral", inner="quart", bw =0.2, cut=1, linewidth=1)
ax.set_xlabel("\n Administrative Regions", size=16)
ax.set_ylabel("Dengue Incidence (log) \n", size=16)
f.suptitle("Monthly dengue incidence in each administrative regions in the Philippines from 2013 to 2018", size=20)

plt.savefig('deng2log.jpg', dpi=150)
plt.show()
```
